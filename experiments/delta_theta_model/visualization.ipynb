````python
import pandas
import matplotlib.pyplot as plt
import os
root = os.getcwd()
````

## Training

We trained for 50 epochs and selected the best model according to our validation loss:

````python
full_df = pandas.read_csv(os.path.join(root, "logs", "starting_data_training.log"))
full_df
````

````python
delta_df = pandas.read_csv(os.path.join(root, "logs", "tf_48_delta_training.log"))
delta_df
````

````python
theta_df = pandas.read_csv(os.path.join(root, "logs", "tf_48_theta_training.log"))
theta_df
````

````python
delta_theta_df = pandas.read_csv(os.path.join(root, "logs", "tf_48_delta_theta_training.log"))
delta_theta_df
````

### Training accuracy

````python
full_acc_columns = [x for x in list(full_df.columns) if "acc" in x]
delta_acc_columns = [x for x in list(delta_df.columns) if "acc" in x]
theta_acc_columns = [x for x in list(theta_df.columns) if "acc" in x]
delta_theta_acc_columns = [x for x in list(delta_theta_df.columns) if "acc" in x]

plt.figure()
full_df[full_acc_columns[0]].plot()
delta_df[delta_acc_columns[0]].plot()
theta_df[theta_acc_columns[0]].plot()
delta_theta_df[delta_theta_acc_columns[0]].plot()
plt.grid(True)
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.ylim([0.5,1.0])
plt.legend(['Full','Delta','Theta','Delta+Theta'])
plt.show()

plt.figure()
full_df[full_acc_columns[1]].plot()
delta_df[delta_acc_columns[1]].plot()
theta_df[theta_acc_columns[1]].plot()
delta_theta_df[delta_theta_acc_columns[1]].plot()
plt.grid(True)
plt.title("Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.ylim([0.5,1.0])
plt.legend(['Full','Delta','Theta','Delta+Theta'])
plt.show()
````

### Training loss

````python
full_acc_columns = [x for x in list(full_df.columns) if "loss" in x]
delta_acc_columns = [x for x in list(delta_df.columns) if "loss" in x]
theta_acc_columns = [x for x in list(theta_df.columns) if "loss" in x]
delta_theta_acc_columns = [x for x in list(delta_theta_df.columns) if "loss" in x]

plt.figure()
full_df[full_acc_columns[0]].plot()
delta_df[delta_acc_columns[0]].plot()
theta_df[theta_acc_columns[0]].plot()
delta_theta_df[delta_theta_acc_columns[0]].plot()
plt.grid(True)
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim([0, 1])
plt.legend(['Full','Delta','Theta','Delta+Theta'])
plt.show()

plt.figure()
full_df[full_acc_columns[1]].plot()
delta_df[delta_acc_columns[1]].plot()
theta_df[theta_acc_columns[1]].plot()
delta_theta_df[delta_theta_acc_columns[1]].plot()
plt.grid(True)
plt.title("Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim([0, 1])
plt.legend(['Full','Delta','Theta','Delta+Theta'])
plt.show()
````

## Evaluation

````python
from scipy import stats
    
with open(os.path.join(root, "evaluations", "starting_data.json")) as fp:
    df1 = pandas.read_json(fp).T
with open(os.path.join(root, "evaluations", "tf_48_delta.json")) as fp:
    df2 = pandas.read_json(fp).T
with open(os.path.join(root, "evaluations", "tf_48_theta.json")) as fp:
    df3 = pandas.read_json(fp).T
with open(os.path.join(root, "evaluations", "tf_48_delta_theta.json")) as fp:
    df4 = pandas.read_json(fp).T
````

Each point in the boxplot is the accuracy for one subject

````python
plt.figure()
plt.grid(color='black', linestyle=':', linewidth=1)
plt.boxplot([df1[["acc"]].T.values[0], df2[["acc"]].T.values[0], df3[["acc"]].T.values[0], df4[["acc"]].T.values[0]])
plt.ylim([0.5,1])
plt.ylabel("Accuracy")
plt.xlabel("Frequency Band")
plt.xticks([1, 2, 3, 4], ["Full", "Delta", "Theta", "Delta+Theta"])
plt.show()
````
